# Test de normalitÃ© Kolmogorov Smirnov
- Si p_value < 0.05, ne suit pas une loi normale

```python
from scipy import stats

def ks_test(col):
    alpha = 0.20
    stat, p = stats.kstest(df[col], 'norm')
    if p < alpha:
        print(f'{col}\n{p}')
        return print('REJETE')
```

```python
for col in polysomno_num:
    ks_test(col)
```
